```python

list_1=( "FPT", "NVL","VND","SSI","DIG","SHB","STB","DXG","TPB","HPG","MWG","TCH","DBC","EIB","GEX","VHM","VPB","HDB","HAG")
list_ck=("SHS","CEO", "PVS", "MBS","HUT","TTH","LAS","TIG","PVC","BVS","IDC","HTP","AAV","DDG","IDJ","VGS","TNG","MST","AMV","VC3","AAH","BSR","AAS","SBS","VGI","DDV","VHG","DRI","VEA", "DGT","C4G","VAB","QTP","BCR","BMS","ICF","BOT","KVC","OIL")

for ck in list_ck:
    params['symbol']=ck
    import sqlite3
    path="D:/tai/csdl.db"
    connection=sqlite3.connect(path)
    cur = connection.cursor()
    table_name = "chung_khoan_" + ck
    create_table_query='''CREATE TABLE IF NOT EXISTS {} (
                    symbol TEXT,
                    date TEXT,
                    close REAL,
                    high REAL,
                    low REAL,
                    open REAL,
                    open_int Real,
                    value REAL,
                    volume REAL,
                    PRIMARY KEY (symbol, date)
                )'''.format(table_name)
    cur.execute(create_table_query)
    result = cur.fetchone()
    connection.commit()  
    response=requests.get('https://svr1.fireant.vn/api/Data/Markets/HistoricalQuotes',headers=headers,params=params)
    if response.status_code==200:
    #print("dung")
        data=response.json()
        for item in data:
        #print(item)
           symbol_price = item['Symbol']
           date_price = item['Date']
           close_price = item['Close']
           high_price = item['High']
           low_price = item['Low']
           open_price = item['Open']
           openInt_price = item['OpenInt']
           value_price = item['Value']
           volume_price = item['Volume']
           print(symbol_price,date_price,close_price,high_price,low_price,open_price,openInt_price,value_price,volume_price)
           data_to_insert = (symbol_price, date_price, close_price, high_price, low_price, open_price, openInt_price, value_price, volume_price)

           cusss='''INSERT INTO {} (symbol, date, close, high, low, open, open_int, value, volume)
                          VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?)'''.format(table_name)
           cur.execute(cusss,data_to_insert)
        connection.commit()  
```
