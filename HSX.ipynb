```python
import sqlite3
import requests
from datetime import datetime
d=[ "MBB","FPT", "NVL","VND","SSI","DIG","SHB","STB","DXG","TPB","HPG","MWG","TCH","DBC","EIB","GEX","VHM","VPB","HDB","HAG"]
def create_table(ck):
    ck="chung_khoan_"+ck
    conn = sqlite3.connect("D:/tai/csdl.db")
    cursor = conn.cursor()
    cu='''CREATE TABLE IF NOT EXISTS {} (
                        symbol TEXT,
                        date TEXT,
                        close REAL,
                        high REAL,
                        low REAL,
                        open REAL,
                        open_int Real,
                        value REAL,
                        volume INTEGER,
                        PRIMARY KEY (symbol, date)
                    )'''.format(ck)
    cursor.execute(cu)
    conn.commit()
    conn.close()

def insert_or_update_data(ck,symbol_price, date_price, close_price, high_price, low_price, open_price, openInt_price, value_price, volume_price):
    ck="chung_khoan_"+ck
    conn = sqlite3.connect("D:/tai/csdl.db")
    cursor = conn.cursor()
    curs='''INSERT OR REPLACE INTO {}
                      (symbol, date, close, high, low, open, open_int, value, volume) 
                      VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?)'''.format(ck)
    cuxx=(symbol_price, date_price, close_price, high_price, low_price, open_price, openInt_price, value_price, volume_price)
    cursor.execute (curs,cuxx)
    conn.commit()
    conn.close()

def crawl_api(ck):
    current_datetime = datetime.now()
    current_date = current_datetime.date()
    params = {
        'symbol': 'MBB',
        'startDate': current_date,
        'endDate': current_date  
    }
    headers = {
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36',
        'Accept':'application/json, text/plain, */*',
        'Accept-Language':'vi,fr-FR;q=0.9,fr;q=0.8,en-US;q=0.7,en;q=0.6',
        'Referer':'https://www.fireant.vn/',
    }
    params["symbol"]=ck
    response = requests.get('https://svr1.fireant.vn/api/Data/Markets/HistoricalQuotes', headers=headers, params=params)    
    if response.status_code == 200:
        data = response.json()
        for item in data:
            symbol_price = item['Symbol']
            date_price = item['Date']
            close_price = item['Close']
            high_price = item['High']
            low_price = item['Low']
            open_price = item['Open']
            openInt_price = item['OpenInt']
            value_price = item['Value']
            volume_price = item['Volume']
            insert_or_update_data(ck,symbol_price, date_price, close_price, high_price, low_price, open_price, openInt_price, value_price, volume_price)
            
            print(symbol_price, date_price, close_price, high_price, low_price, open_price, openInt_price, value_price, volume_price)
    else:
        print("Lỗi khi truy cập API:", response.status_code)


while True:
    for ck in d:
        create_table(ck)
        crawl_api(ck) 
    time.sleep(30)
```
